```python
import pandas as pd

# Ganti path sesuai dengan lokasi dataset di Kaggle Anda
file_path = "/kaggle/input/rossmann-store-sales/train.csv"

# Membaca dataset ke dalam DataFrame
df = pd.read_csv(file_path)

# Menampilkan beberapa baris pertama dari DataFrame
df.head()
```

## 2. Visualisasi dengan BarPlot
### Plot jumlah penjualan pada setiap hari

ChatGPT Prompt:
Berikan saya code untuk plot kolom 'Customer' terhadap kolom 'day of week' dengan barplot menggunakan data pada df

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Mengelompokkan data berdasarkan 'day of week' dan menghitung rata-rata 'Customer' untuk setiap kelompok
customer_by_day = df.groupby('DayOfWeek')['Customers'].mean().reset_index()

# Membuat barplot
plt.figure(figsize=(10, 6))
sns.barplot(x='DayOfWeek', y='Customers', data=customer_by_day, palette='viridis')
plt.title('Rata-rata Customer per Hari dalam Seminggu')
plt.xlabel('Day of Week')
plt.ylabel('Rata-rata Customer')
plt.xticks(rotation=45)
plt.show()
```

ChatGPT Prompt: 
Berikan saya code untuk plot kolom 'Sales' terhadap kolom 'DayofWeek'  dengan barplot menggunakan library seaborn

```python
# Membuat barplot
plt.figure(figsize=(10, 6))
sns.barplot(x='DayOfWeek', y='Sales', data=df, palette='viridis')
plt.title('Penjualan per Hari dalam Seminggu')
plt.xlabel('Hari dalam Seminggu')
plt.ylabel('Penjualan')
plt.xticks(rotation=45)
plt.show()
```

ChatGPT Prompt: Tambahkan kode untuk mengatur variasi palet seaborn agar warna pada grafik berubah

```python
# Mengatur tema Seaborn
sns.set_style("whitegrid")

# List palet warna dengan variasi yang berbeda
palet_warna = sns.color_palette("Set2")

# Membuat barplot
plt.figure(figsize=(10, 6))
sns.barplot(x='DayOfWeek', y='Sales', data=df, palette=palet_warna)
plt.title('Penjualan per Hari dalam Seminggu dengan Variasi Warna')
plt.xlabel('Hari dalam Seminggu')
plt.ylabel('Penjualan')
plt.xticks(rotation=45)
plt.show()
```

## 5. Variasi Seaborn Style 

ChatGPT Prompt: 
Berikan saya code untuk menampilkan visualisasi dataset rossman store sales, kolom 'sales', menggunakan boxplot dengan variasi:
Seaborn Style dan Color Palettes

```python
# Mengatur tema Seaborn
sns.set_style("whitegrid")

# Memilih palet warna Seaborn
palet_warna = "husl"  # Anda dapat mengganti dengan palet warna yang lain sesuai keinginan

# Membuat boxplot kolom 'Sales' dengan variasi gaya dan palet warna
plt.figure(figsize=(10, 6))
sns.boxplot(x='Sales', data=df, palette=palet_warna)
plt.title('Boxplot Penjualan (Sales) dengan Seaborn Style dan Color Palette')
plt.xlabel('Penjualan (Sales)')
plt.show()
```

ChatGPT Prompt: Berikan saya kode untuk menampilkan visualisasi kolom 'sales' dengan berbagai variasi seaborn color palletes

```python
# List palet warna yang berbeda
palet_warna = ["deep", "muted", "pastel", "bright", "dark", "colorblind"]

# Membuat subplots untuk menampilkan boxplot dengan berbagai palet warna
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle("Visualisasi Kolom 'Sales' dengan Berbagai Palet Warna Seaborn", fontsize=16)

for i, palet in enumerate(palet_warna):
    row = i // 3
    col = i % 3
    ax = axes[row, col]
    
    sns.boxplot(x='Sales', data=df, palette=palet, ax=ax)
    ax.set_title(f'Palet Warna: {palet}')
    ax.set_xlabel('Penjualan (Sales)')

# Mengatur tampilan subplot
for ax in axes.flat:
    ax.label_outer()

plt.tight_layout()
plt.show()
```

ChatGPT Prompt: Berikan saya kode untuk membuat beberapa box plot dengan berbagai tema Seaborn yang berbeda untuk kolom 'Sales' dari dataset saya. Saya ingin melihat bagaimana tampilan box plot berubah dengan berbagai tema Seaborn.

```python
# List tema Seaborn yang berbeda
themes = ["darkgrid", "whitegrid", "dark", "white", "ticks"]

# Membuat subplot untuk menampilkan box plot dengan berbagai tema
plt.figure(figsize=(15, 5))
for i, theme in enumerate(themes):
    plt.subplot(1, len(themes), i + 1)
    
    # Mengatur tema Seaborn
    sns.set_theme(style=theme)
    
    # Membuat box plot
    sns.boxplot(x='Sales', data=df)
    
    plt.xlabel('Sales')
    plt.title(f'Box Plot with {theme.capitalize()} Theme')

plt.tight_layout()
plt.show()
```

## 6. Plot Regresi Linear

ChatGPT Prompt: Berikan kode untuk menampilkan plot regresi linear pada data rossman store sales dengan library seaborn

```python
# Membuat plot regresi linear
plt.figure(figsize=(10, 6))
sns.regplot(x='Customers', y='Sales', data=df, scatter_kws={"color": "blue"}, line_kws={"color": "red"})
plt.title('Plot Regresi Linear antara Customers dan Sales')
plt.xlabel('Customers')
plt.ylabel('Sales')
plt.show()
```
