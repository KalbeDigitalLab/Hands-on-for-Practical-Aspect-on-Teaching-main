  `plt.figure(figsize=(12, 6))`

  `plt.plot(df.index, y_pred, color='r', label='Linear Regression')`

```python
from sklearn.linear_model import LinearRegression

# Menyiapkan data untuk regresi linear
X = df.index.values.astype(float).reshape(-1, 1)
y = df['#Passengers'].values

# Melakukan regresi linear
model = LinearRegression()
model.fit(X, y)

# Memprediksi nilai menggunakan model regresi linear
y_pred = model.predict(X)

# Membuat plot garis untuk data asli dan hasil regresi
plt.figure(figsize=(12, 6))
plt.scatter(df.index, df['#Passengers'], color='b', label='Actual Data')
plt.plot(df.index, y_pred, color='r', label='Linear Regression')
plt.title('Linear Regression of Time Series Data')
plt.xlabel('Years')
plt.ylabel('#Passengers')
plt.legend()
plt.show()
```

```python
# Menghitung rata-rata bergerak sederhana (SMA)
sma = df.rolling(window=12).mean()

# Menghitung rata-rata bergerak tertimbang (WMA)
weights = list(range(1, 13))
wma = df.rolling(window=12).apply(lambda prices: (prices * weights).sum() / sum(weights), raw=True)

# Menghitung rata-rata bergerak eksponensial (EMA)
ema = df.ewm(span=12, adjust=False).mean()

# Membuat plot garis untuk SMA, WMA, dan EMA
plt.figure(figsize=(12, 6))
plt.plot(df, label='Original')
plt.plot(sma, label='SMA', linestyle='dashed')
plt.plot(wma, label='WMA', linestyle='dotted')
plt.plot(ema, label='EMA', linestyle='dashdot')
plt.title('Simple Moving Average, Weighted Moving Average, and Exponential Moving Average')
plt.xlabel('Years')
plt.ylabel('Values')
plt.legend()
plt.show()
```

Catatan/Insights/Penjelasan:
* **SMA** adalah metode perhitungan moving average yang paling sederhana. Ini adalah rata-rata aritmatika dari harga selama periode waktu tertentu. SMA memberikan bobot yang sama untuk setiap harga dalam rentang waktu yang dipilih.
* **WMA** memberi bobot yang berbeda pada harga-harga yang berbeda dalam rentang waktu yang dipilih. Harga-harga terbaru diberi bobot lebih besar daripada harga-harga yang lebih lama dalam rentang waktu tersebut. Ini memungkinkan WMA untuk bereaksi lebih cepat terhadap perubahan harga terbaru.
* **EMA** memberi bobot yang lebih besar pada harga terbaru daripada harga yang lebih lama. Ini memberikan perhatian yang lebih besar pada pergerakan harga terkini, sehingga memberikan sinyal tren yang lebih cepat daripada SMA. Formula EMA melibatkan penggunaan konstan penghalus yang mengurangi bobot harga lama seiring waktu. EMA dapat dianggap sebagai gabungan dari SMA dan WMA dengan bobot yang terus berkurang seiring berjalannya waktu.

```python
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# ARIMA modeling
order = (30, 1, 5)  # Example order, you might need to tune this
arima_model = ARIMA(df['#Passengers'], order=order)
arima_result = arima_model.fit()

# Plot original time series, ACF, PACF, and ARIMA forecast
plt.figure(figsize=(14, 10))

# Original time series plot
plt.subplot(2, 2, 1)
plt.plot(df['#Passengers'], label='Jumlah Penumpang')
plt.title('Original Time Series')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Penumpang')
plt.legend()

# Plot ACF
plt.subplot(2, 2, 2)
plot_acf(df['#Passengers'], lags=40, ax=plt.gca(), title='Autocorrelation Function (ACF)')

# Plot PACF
plt.subplot(2, 2, 3)
plot_pacf(df['#Passengers'], lags=40, ax=plt.gca(), title='Partial Autocorrelation Function (PACF)')

# Plot ARIMA forecast for the next 10 years
forecast_steps = 120
arima_forecast = arima_result.get_forecast(steps=forecast_steps)
arima_mean = arima_forecast.predicted_mean
plt.subplot(2, 2, 4)
plt.plot(df['#Passengers'], label='Jumlah Penumpang')
plt.plot(arima_mean.index, arima_mean, label=f'ARIMA({order[0]},{order[1]},{order[2]}) Forecast', linestyle='dashed')
plt.title('ARIMA Forecast (10 Years Ahead)')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Penumpang')
plt.legend()

plt.tight_layout()
plt.show()
```
