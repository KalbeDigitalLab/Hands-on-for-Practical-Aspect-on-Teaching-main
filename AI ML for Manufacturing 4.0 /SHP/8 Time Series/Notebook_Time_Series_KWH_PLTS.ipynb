Beri kode untuk menghapus outliers

```python
# Fungsi untuk menghapus outlier berdasarkan IQR
def remove_outliers(df, col_name):
    Q1 = df[col_name].quantile(0.25)
    Q3 = df[col_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[col_name] > lower_bound) & (df[col_name] < upper_bound)]
    return df

# Hapus outlier dari df_train
df_train = remove_outliers(df_train, 'PLTS2 KWH')
```

Beri kode untu membuat sebuah fungsi bernama windowed_dataset yang mengambil empat argumen, yaitu series (data input), window_size (ukuran jendela), batch_size (ukuran batch), dan shuffle_buffer (buffer pengacakan). Fungsi ini digunakan untuk membuat dataset berdasarkan jendela geser pada data input series dengan ukuran window_size, mengacak dataset menggunakan shuffle_buffer, dan mengembalikan dataset dalam bentuk batch dengan ukuran batch_size yang siap digunakan untuk pelatihan model.

```python
def windowed_dataset(series, window_size, batch_size, shuffle_buffer):
    ds = tf.data.Dataset.from_tensor_slices(series)
    ds = ds.window(window_size + 1, shift=1, drop_remainder=True)
    ds = ds.flat_map(lambda w: w.batch(window_size + 1))
    ds = ds.shuffle(shuffle_buffer)
    ds = ds.map(lambda w: (w[:-1], w[-1]))
    return ds.batch(batch_size).prefetch(1)
```

```python
train_set = windowed_dataset(kwh_scale_train, window_size=60, batch_size=100, shuffle_buffer=1000)

val_set = windowed_dataset(kwh_scale_val, window_size=60, batch_size=100, shuffle_buffer=1000)

model = tf.keras.models.Sequential([
    tf.keras.layers.LSTM(64, return_sequences=True, input_shape=(None, 1)),
    tf.keras.layers.LSTM(64),
    tf.keras.layers.Dense(30, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(1)
])
```

Beri kode untuk membuat kelas myCallback yang merupakan turunan dari tf.keras.callbacks.Callback. Fungsi on_epoch_end dipanggil di akhir setiap epoch selama pelatihan model. Di dalam fungsi ini, metrik 'val_mae' dievaluasi dari logs, yang merupakan Mean Absolute Error (MAE) pada data validasi. Jika MAE lebih kecil dari 0.09, maka kondisi ini terpenuhi. Dalam kasus ini, pesan dicetak untuk memberitahukan bahwa model telah mencapai target MAE kurang dari 0.09, dan pelatihan model dihentikan dengan menetapkan atribut stop_training menjadi True.

Kemudian beri kode untuk membuat objek callback dari kelas myCallback

```python
class myCallback(tf.keras.callbacks.Callback):
  # Fungsi ini dipanggil di akhir setiap epoch selama pelatihan model
  def on_epoch_end(self, epoch, logs={}):
    # Mengevaluasi metrik 'val_mae' dari logs yang merupakan Mean Absolute Error (MAE) pada data validasi
    # Jika MAE lebih kecil dari 0.09, maka kondisi ini terpenuhi
    if(logs.get('val_mae')<0.09):
      # Mencetak pesan bahwa model telah mencapai target MAE kurang dari 0.09
      print("MAE model telah mencapai < 10%! ")
      # Menghentikan pelatihan model dengan menetapkan atribut stop_training menjadi True
      self.model.stop_training = True

# Membuat objek callback dari kelas myCallback yang telah didefinisikan sebelumnya
callbacks = myCallback()
```

Prompt:

Beri kode yang mendefinisikan optimizer SGD (Stochastic Gradient Descent) dengan learning rate 0.1 dan momentum 0.9. Kemudian mengompilasi model dengan fungsi loss Huber, optimizer yang telah didefinisikan, dan metrik MAE. Terakhir, melatih model dengan data pelatihan dan data validasi dengan jumlah epochs 50, dan menggunakan callbacks.

```python
# Mendefinisikan optimizer SGD (Stochastic Gradient Descent) dengan learning rate 0.1 dan momentum 0.9
optimizer = tf.keras.optimizers.SGD(learning_rate=1.0000e-01, momentum=0.9)

# Mengompilasi model dengan fungsi loss Huber, optimizer yang sudah didefinisikan, dan metrik MAE
model.compile(loss=tf.keras.losses.Huber(),
              optimizer=optimizer,
              metrics=["mae"])

# Melatih model dengan data pelatihan dan data validasi
history = model.fit(train_set, epochs=50, callbacks=[callbacks], validation_data=val_set)
```
