Import library yang diperlukan:

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from lifelines import KaplanMeierFitter, NelsonAalenFitter
```

Siapkan kolom untuk waktu (time) dan status (event):

```python
time_col = 'Tool wear [min]'
event_col = 'Failure Type'
```

Konversi kolom waktu ke bentuk integer (jika belum):

```python
df[time_col] = df[time_col].astype(int)
```

Inisialisasi objek Kaplan-Meier:

```python
kmf = KaplanMeierFitter()
```

Pisahkan data berdasarkan kategori pada kolom Failure Type:

```python
for failure_type in df[event_col].unique():
    data = df[df[event_col] == failure_type]

    kmf.fit(data[time_col], event_observed=None, label=failure_type)
    kmf.plot_survival_function()

plt.title('Kaplan-Meier Survival Curve by Failure Type')
plt.xlabel('Time (minutes)')
plt.ylabel('Survival Probability')
plt.show()
```

Buat Cumulative Hazard Plot menggunakan Nelson-Aalen Estimator:

```python
naf = NelsonAalenFitter()

for failure_type in df[event_col].unique():
    data = df[df[event_col] == failure_type]

    naf.fit(data[time_col], event_observed=None, label=failure_type)
    naf.plot_cumulative_hazard()

plt.title('Cumulative Hazard Curve by Failure Type')
plt.xlabel('Time (minutes)')
plt.ylabel('Cumulative Hazard')
plt.show()
```

Buat Log-Log Survival Plot:

```python
sns.set(style="whitegrid")

for failure_type in df[event_col].unique():
    data = df[df[event_col] == failure_type]

    kmf.fit(data[time_col], event_observed=None, label=failure_type)

    # Get the survival function values
    survival_prob = 1 - kmf.survival_function_[failure_type]

    # Plot Log-Log Survival Curve
    plt.plot(np.log(kmf.survival_function_.index), np.log(survival_prob), label=failure_type)

plt.title('Log-Log Survival Curve by Failure Type')
plt.xlabel('Log(Time)')
plt.ylabel('Log(Survival Probability)')
plt.legend()
plt.show()
```

Buat Survival Contour Plot (jika ingin):

```python
sns.set(style="whitegrid")

g = sns.histplot(x=time_col, y=event_col, data=df, stat="density", common_norm=False, bins=30)

plt.show()
```
