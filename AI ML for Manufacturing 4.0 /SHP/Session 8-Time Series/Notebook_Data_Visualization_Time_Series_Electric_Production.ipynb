Membaca File CSV dan Menampilkan Dataframe:

Kode ini membaca file CSV dari Google Drive ke dalam sebuah dataframe pandas, dan kemudian menampilkan beberapa baris pertama dari dataframe.

```python
import pandas as pd

# Ganti path dengan path file CSV di Google Drive Anda
file_path = '/content/drive/MyDrive/Dataset/Electric_Production.csv'

# Membaca file CSV ke dalam dataframe
df = pd.read_csv(file_path)

# Menampilkan beberapa baris pertama dari dataframe
df.head()
```

Visualisasi Scatter Plot:

Kode ini membuat scatter plot dari data dengan sumbu x sebagai tanggal ('DATE') dan sumbu y sebagai nilai ('Value').

```python
import matplotlib.pyplot as plt

# Konversi kolom 'Date' menjadi format datetime
df['DATE'] = pd.to_datetime(df['DATE'])

# Menyiapkan plot
plt.figure(figsize=(12, 6))

# Membuat scatter plot
plt.scatter(df['DATE'], df['Value'], color='b', alpha=0.5)

# Menambahkan judul dan label sumbu
plt.title('Scatter Plot of Time Series Data', fontsize=16)
plt.xlabel('DATE', fontsize=12)
plt.ylabel('Value', fontsize=12)

# Menampilkan plot
plt.show()
```

Visualisasi Time Series Plot:
Kode ini mengatur kolom 'DATE' sebagai indeks dari dataframe dan kemudian membuat plot time series dari data.

```python
df.set_index('DATE',inplace=True)
df.head()
```

```python
df.plot()
plt.grid()
plt.show()
```

Analisis Dekomposisi Musiman:

Kode ini mengimpor fungsi dekomposisi musiman dari pustaka statsmodels, melakukan dekomposisi musiman dari data time series, dan kemudian menampilkan plot untuk komponen yang diamati dan tren.

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
decompose = seasonal_decompose(df,model='additive')
decompose.plot(observed=True,trend=True)
plt.grid()
plt.show()
```
