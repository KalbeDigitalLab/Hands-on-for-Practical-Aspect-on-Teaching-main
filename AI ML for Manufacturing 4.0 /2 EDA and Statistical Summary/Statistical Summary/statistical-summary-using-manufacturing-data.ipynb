## Load the dataset

ChatGPT Prompt: I have the dataset on kaggle. After adding the data to kaggle notebook, write me a code to load the dataset

```python
import pandas as pd

# Load the dataset into a pandas DataFrame
df = pd.read_csv('/kaggle/input/manufacturing-cost/EconomiesOfScale.csv')

# Display the first few rows of the DataFrame
print(df.head())
```

## Perform EDA and Statistical Summary

ChatGPT Prompt: Give me a code to perform EDA with statistical summary on that data

```python
# Display the first few rows of the DataFrame
print("Preview of the dataset:")
print(df.head())

# Summary statistics
print("\nSummary Statistics:")
print(df.describe())

# Info about columns and data types
print("\nInfo about columns and data types:")
print(df.info())
```

## Calculate Quantiles

ChatGPT Prompt: Give me the code to calculate quantiles on that data

```python
# Specify the column for which you want to calculate quantiles
column_name = 'Number of Units'

# Specify the quantiles you want to calculate
quantiles_to_calculate = [0.25, 0.50, 0.75]

# Calculate the quantiles
quantiles_values = df['Number of Units'].quantile(quantiles_to_calculate)

# Display the calculated quantiles
print("Calculated Quantiles for", 'Number of Units')
print(quantiles_values)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a box plot using Seaborn

sns.boxplot(y='Number of Units', data=df)
plt.title("Box Plot of " + 'Number of Units')
plt.ylabel('Number of Units')

# Add horizontal lines for the calculated quantiles
for quantile in quantiles_values:
    plt.axhline(y=quantile, color='r', linestyle='--', label=f'{quantile:.2f}')

# Display the plot
plt.legend()
plt.show()
```

## Skewness and Kurtosis

ChatGPT Prompt: Give me the code to show skewness and curtosis

```python
# Specify the column for which you want to calculate skewness and kurtosis
column_name = 'Manufacturing Cost'

# Calculate skewness and kurtosis
skewness = df['Manufacturing Cost'].skew()
kurtosis = df['Manufacturing Cost'].kurtosis()

# Display the calculated skewness and kurtosis
print("Skewness for", column_name, ":", skewness)
print("Kurtosis for", column_name, ":", kurtosis)
```

## Covariance & Scatterplot

ChatGPT Prompt: give me the code to show skewness and curtosis

```python
# Specify the two columns for which you want to calculate covariance and create a scatter plot
column_x = 'Number of Units'
column_y = 'Manufacturing Cost'

# Calculate covariance between the two columns
covariance = df[column_x].cov(df[column_y])

# Create a scatter plot using Seaborn
sns.scatterplot(x=column_x, y=column_y, data=df)

# Set plot title and labels
plt.title("Scatter Plot")
plt.xlabel(column_x)
plt.ylabel(column_y)

# Display covariance in the plot
plt.text(df[column_x].min(), df[column_y].max(), f'Covariance: {covariance:.2f}', fontsize=12, color='red')

# Display the plot
plt.show()
```

## Bivariate Analysis

ChatGPT Prompt: Write Python codes to perform bivariate analysis using Pearson, Spearman and Kendall correlation coefficient and interpret the results.

```python
import pandas as pd
import numpy as np
import scipy.stats



# Calculate Pearson correlation coefficient
pearson_corr = df[column_x].corr(df[column_y])

# Calculate Spearman correlation coefficient
spearman_corr = df[column_x].corr(df[column_y], method='spearman')

# Calculate Kendall correlation coefficient
kendall_corr = df[column_x].corr(df[column_y], method='kendall')

# Print the correlation coefficients
print("Pearson Correlation Coefficient:", pearson_corr)
print("Spearman Correlation Coefficient:", spearman_corr)
print("Kendall Correlation Coefficient:", kendall_corr)

# Interpretation
if abs(pearson_corr) > 0.7:
    pearson_interpretation = "Strong"
elif abs(pearson_corr) > 0.3:
    pearson_interpretation = "Moderate"
else:
    pearson_interpretation = "Weak"

if abs(spearman_corr) > 0.7:
    spearman_interpretation = "Strong"
elif abs(spearman_corr) > 0.3:
    spearman_interpretation = "Moderate"
else:
    spearman_interpretation = "Weak"

if abs(kendall_corr) > 0.7:
    kendall_interpretation = "Strong"
elif abs(kendall_corr) > 0.3:
    kendall_interpretation = "Moderate"
else:
    kendall_interpretation = "Weak"

print("\nInterpretation:")
print(f"Pearson Correlation is {pearson_interpretation}.")
print(f"Spearman Correlation is {spearman_interpretation}.")
print(f"Kendall Correlation is {kendall_interpretation}.")
```
