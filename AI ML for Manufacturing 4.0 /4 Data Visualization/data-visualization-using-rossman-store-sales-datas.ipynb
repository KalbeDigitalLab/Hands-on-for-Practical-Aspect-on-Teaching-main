## 5. Variasi Seaborn Style 

ChatGPT Prompt: 
Berikan saya code untuk menampilkan visualisasi dataset rossman store sales, kolom 'sales', menggunakan boxplot dengan variasi:
Seaborn Style, Context Functions, dan Color Palettes

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set style Seaborn
sns.set(style="whitegrid")  # Gaya whitegrid

# Set konteks Seaborn
sns.set_context("notebook", font_scale=1.2)  # Konteks notebook dengan peningkatan ukuran font

# Pilih palet warna
color_palette = sns.color_palette("Set2")  # Gunakan palet warna Set2

# Membuat box plot untuk kolom 'Sales'
plt.figure(figsize=(8, 6))  # Mengatur ukuran gambar

sns.boxplot(x='Sales', data=df, palette=color_palette)  # Membuat box plot dengan palet warna yang dipilih
plt.xlabel('Sales')  # Memberi label sumbu x
plt.title('Box Plot of Sales')  # Memberi judul plot

plt.show()  # Menampilkan plot
```

ChatGPT Prompt: Berikan saya kode untuk menampilkan visualisasi kolom 'sales' dengan berbagai variasi seaborn color palletes

```python
# Pilih beberapa palet warna Seaborn yang berbeda
palettes = ["deep", "muted", "pastel", "dark", "colorblind"]

# Buat subplot untuk menampilkan box plot dengan berbagai palet warna
plt.figure(figsize=(15, 5))

for i, palette in enumerate(palettes):
    plt.subplot(1, len(palettes), i + 1)
    
    # Membuat box plot dengan palet warna yang dipilih
    sns.boxplot(x='Sales', data=df, palette=palette)
    
    plt.xlabel('Sales')
    plt.title(f'Box Plot with {palette.capitalize()} Palette')

plt.tight_layout()
plt.show()
```

ChatGPT Prompt: Berikan saya kode untuk menampilkan visualisasi kolom 'sales' dengan berbagai variasi seaborn themes

```python
# Pilih beberapa tema Seaborn yang berbeda
themes = ["darkgrid", "whitegrid", "dark", "white", "ticks"]

# Buat subplot untuk menampilkan box plot dengan berbagai tema
plt.figure(figsize=(15, 5))

for i, theme in enumerate(themes):
    plt.subplot(1, len(themes), i + 1)
    
    # Mengatur tema Seaborn
    sns.set_theme(style=theme)
    
    # Membuat box plot
    sns.boxplot(x='Sales', data=df)
    
    plt.xlabel('Sales')
    plt.title(f'Box Plot with {theme.capitalize()} Theme')

plt.tight_layout()
plt.show()
```

## 6. Plot Regresi Linear

ChatGPT Prompt: Berikan kode untuk menampilkan plot regresi linear pada data rossman store sales dengan library seaborn

```python
# Menggunakan lmplot untuk menampilkan plot regresi linear
plt.figure(figsize=(10, 6))

sns.lmplot(x='Sales', y='Customers', data=df, aspect=2, height=6)
plt.xlabel('Sales')
plt.ylabel('Customers')
plt.title('Linear Regression Plot of Sales vs Customers')

# Menampilkan plot
plt.show()
```
