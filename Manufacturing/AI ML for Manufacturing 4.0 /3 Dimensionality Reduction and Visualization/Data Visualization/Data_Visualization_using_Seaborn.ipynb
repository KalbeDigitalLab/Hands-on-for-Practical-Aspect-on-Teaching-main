# Step 5. Data Visualization using Seaborn

Prompt: Give me a python code to import the necessary libraries and loads the 'train.csv' dataset into a Pandas DataFrame. Display, the first few rows of the DataFrame.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('train.csv')
df.head()
```

## Plot Distribution of Target (y) using Histogram

Prompt: Write me a code to plot histogram distribution of 'y' using seaborn

```python
# Distribution of 'y' - the time taken for testing
sns.histplot(df['y'], bins=20, kde=False, color='blue')
plt.xlabel('Time Taken for Testing')
plt.ylabel('Frequency')
plt.title('Distribution of Testing Time')
plt.show()
```

# Data Visualization using boxplot

Prompt: Write me a code to visualize feature 'X0' vs target 'y' using seaborn boxplot on Mercedes-Benz greener dataset

```python
# Boxplot for 'X0' vs 'y' (considering 'X0' is a categorical feature)
sns.boxplot(x='X0', y='y', data=df)
plt.xlabel('X0')
plt.ylabel('Time Taken for Testing')
plt.title('Boxplot of X0 vs Testing Time')
plt.xticks(rotation=90)  # Rotate x labels for better visibility
plt.show()
```

```python
sns.boxplot(x='y', data=df)
plt.show()
```

## Visualize using Different Seaborn themes

Prompt: Give me a code to visualize boxplot using different seaborn themes

```python
# List of Seaborn themes to apply
seaborn_themes = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']

for theme in seaborn_themes:
    sns.set_style(theme)

    # Create a boxplot
    sns.boxplot(x='X0', y='y', data=df)

    # Labels and Title
    plt.xlabel('X0')
    plt.ylabel('Time Taken for Testing (y)')
    plt.title(f'Boxplot of X0 vs Time Taken for Testing (y) - Theme: {theme}')

    # Show the plot
    plt.show()
```

## Visualize using Different Seaborn Color Palettes

Prompt: Give me a code to visualize using different seaborn color palettes

```python
# List of Seaborn color palettes to apply
seaborn_palettes = ['deep', 'muted', 'pastel', 'dark', 'colorblind']

for palette in seaborn_palettes:
    # Create a boxplot with a different color palette
    sns.boxplot(x='X0', y='y', data=df, palette=palette)

    # Labels and Title
    plt.xlabel('X0')
    plt.ylabel('Time Taken for Testing (y)')
    plt.title(f'Boxplot of X0 vs Time Taken for Testing (y) - Palette: {palette}')

    # Show the plot
    plt.show()
```

## Visualize using Histogram and Density Plot

Prompt: Give me a code to visualize using histogram, density plot, and histoplot

```python
# Create a histogram
plt.hist(df['y'], bins=20, color='blue', edgecolor='black')
plt.xlabel('Time Taken for Testing (y)')
plt.title('Histogram of Time Taken for Testing')
plt.show()
```

```python
# Create a density plot (KDE)
sns.kdeplot(df['y'], shade=True, color='green')
plt.xlabel('Time Taken for Testing (y)')
plt.title('Density Plot of Time Taken for Testing')
plt.show()
```

```python
# Create a combined histogram and density plot
sns.histplot(df['y'], bins=20, kde=True, color='red')
plt.xlabel('Time Taken for Testing (y)')
plt.title('Combined Histogram and Density Plot of Time Taken for Testing')
plt.show()
```

## Scatter Plot

Prompt: Give me a code to create scatter plot

```python
# Create a scatter plot
sns.scatterplot(x='X1', y='y', data=df, hue='X1', palette='viridis')
plt.xlabel('X1')
plt.ylabel('Time Taken for Testing (y)')
plt.title('Scatter Plot of X1 vs Time Taken for Testing')
plt.show()
```
